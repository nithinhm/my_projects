```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service as ChromeService
from bs4 import BeautifulSoup
import pandas as pd

batch = '22'  # enter batch year as in USN (4th and 5th character)
branch = 'CS'  # enter branch as in USN (6th and 7th character)
final_number = 1  # enter the last number of this branch's last student as in USN (do not include padded zeros)

df = pd.DataFrame()

options = Options()
options.unhandled_prompt_behavior = 'ignore'

service = ChromeService(executable_path='/chromedriver.exe')
driver = webdriver.Chrome(service=service, options=options)

url = 'https://results.vtu.ac.in/JFEcbcs23/index.php'

driver.get(url)

skipped_usns = []

usn = f'1AM{batch}{branch}{1:03d}'

usn_bar = driver.find_element(By.NAME, 'lns')
usn_bar.send_keys(usn)

captcha_bar = driver.find_element(By.XPATH, '//*[@id="raj"]/div[2]/div[1]/label')
captcha_bar.click()

try:
    WebDriverWait(driver, 100).until_not(lambda d: d.find_element(By.NAME, 'lns'))

except:
    alert = WebDriverWait(driver, 1).until(EC.alert_is_present())
    print(f'Error for {usn} because {alert.text}')

    if alert.text == 'University Seat Number is not available or Invalid..!':
        skipped_usns.append(usn)
        alert.accept()

student_name = driver.find_element(By.XPATH,
                                    '//*[@id="dataPrint"]/div[2]/div/div/div[2]/div[1]/div/div/div[1]/div/table/tbody/tr[2]/td[2]').text.split(':')[1].strip()

student_usn = driver.find_element(By.XPATH,
                                    '//*[@id="dataPrint"]/div[2]/div/div/div[2]/div[1]/div/div/div[1]/div/table/tbody/tr[1]/td[2]').text.split(':')[1].strip()

assert student_usn == usn, 'entered usn and obtained usn don\'t match'

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

marks_data = soup.find('div', class_='divTableBody')

rows = marks_data.find_all('div', class_='divTableRow')

data = []
for row in rows:
    # Extract the cell values from each row
    cells = row.find_all('div', class_='divTableCell')
    data.append([cell.text.strip() for cell in cells])

# Convert the data into a pandas DataFrame
df_temp = pd.DataFrame(data[1:], columns=data[0])

# # Extract the student name and USN
# student_name = df_temp.loc[0, 'Student Name'].split(':')[1].strip()
# student_usn = df_temp.loc[0, 'University Seat Number'].split(':')[1].strip()

# assert student_usn == usn, 'Entered USN and obtained USN don\'t match'

# df_temp.drop(['Student Name', 'University Seat Number'], axis=1, inplace=True)
# df_temp.set_index('Subject Code', inplace=True)

# df_temp = df_temp.T
# df_temp.insert(0, 'Student Name', student_name)

# df = pd.concat([df, df_temp])

# print(f'Data successfully collected for {usn}')

driver.quit()

# print(f'These USNs were skipped: {skipped_usns}')

# df.index.rename('USN', inplace=True)

# df.to_excel(f'20{batch} {branch} temp VTU results.xlsx')

df_temp
```
